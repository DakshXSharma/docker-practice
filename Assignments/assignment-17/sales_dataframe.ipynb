```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
spark = SparkSession.builder \
    .appName("SalesDataFrameApp") \
    .getOrCreate()
```

```python
df = spark.read.csv(
    "sales.csv",
    header=True,
    inferSchema=True
)
```

```python
df.show()
```

```python
sorted_df = df.orderBy(
    col("sales").desc()
)
```

```python
sorted_df.show()
```

```python
top3 = sorted_df.limit(3)
```

```python
top3.show()
```

```python
high_sales = df.filter(
    col("sales") > 80000
)
```

```python
high_sales.show()
```

```python
high_sales.write.mode("overwrite").csv(
    "output/high_sales_products",
    header=True
)

print("CSV file saved successfully")
```

```python
spark.stop()
```
