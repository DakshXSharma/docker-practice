```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("EmployeeRDD").getOrCreate()
sc = spark.sparkContext
```

```python
lines = sc.textFile("employees.csv")
header = lines.first()
data = lines.filter(lambda line: line != header)
employees = data.map(lambda x: x.split(","))
```

```python
sorted_employees = employees.sortBy(
    lambda x: int(x[3]),
    ascending=False
)
sorted_employees.collect()
```

```python
dept_salary = employees.map(
    lambda x: (x[2], int(x[3]))
).reduceByKey(lambda a, b: a + b)
dept_salary.collect()
```

```python
top3 = sorted_employees.take(3)
top3
```

```python
with open("output/top3.txt", "w") as file:
    for emp in top3:
        file.write(str(emp) + "\n")
print("Top 3 employees saved successfully")
```

```python
spark.stop()
```
